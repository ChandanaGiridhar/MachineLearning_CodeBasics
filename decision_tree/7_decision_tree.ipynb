```python
import pandas as pd
df = pd.read_csv("/content/salaries.csv")
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis = 'columns')
inputs
```

```python
targets = df['salary_more_then_100k']
targets
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis = 'columns')
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs_n,targets)
```

```python
model.score(inputs_n, targets), model.predict([[2,2,1]])
```

```python
import pandas as pd
df1 = pd.read_csv("/content/titanic.csv")
df1.head()
```

```python
input1 = df1.drop(['Survived','PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
target=df1['Survived']
```

```python
input1.Sex = input1.Sex.map({'male':1,'female':2})
```

```python
input1.Age[:10]
```

```python
input1.Age = input1.Age.fillna(input1.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(input1,target,test_size = 0.1)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
