```python
import pandas as pd
df = pd.read_csv("/content/homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
#merge df and dummies
merged = pd.concat([df,dummies],axis = 'columns')
merged
```

```python
#Drop town column - does not work on regression model
final = merged.drop(['town','west windsor'],axis =
                  'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis = 'columns')
X
```

```python
Y = final.price
Y
```

```python
model.fit(X,Y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,Y)
```
