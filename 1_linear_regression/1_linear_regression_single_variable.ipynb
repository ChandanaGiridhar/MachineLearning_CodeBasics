```python
import pandas as pd
import numpy as np #np, pd is alias
import matplotlib.pyplot as plt
from sklearn import linear_model as lr
```

```python
# creating_dataframe
df = pd.read_csv("/content/homeprices.csv")
df # printing dataframe
df.head(3) # FETCH first 3 records
df.tail(3) # last 3 records
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color='red',marker='^') # making scatter plot
plt.xlabel('area("sqft")', fontsize = 10) # giving x and y lables to the graph
plt.ylabel('price("USD")', fontsize = 10)
```

```python
reg = lr.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#validate
#y = mx+c
y = (reg.coef_*5000)+reg.intercept_
y
```

```python
reg.predict([[10000]])
```

```python
new_y = (reg.coef_*10000)+reg.intercept_
new_y
```
